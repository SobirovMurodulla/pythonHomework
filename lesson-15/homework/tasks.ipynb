Task 1: Basic Plotting.

    Task: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.



```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10, 400)


y = f(x)

plt.plot(x, y, label='$f(x) = x^2 - 4x + 4$')


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function $f(x) = x^2 - 4x + 4$')
plt.legend()

plt.grid(True)
plt.show()
```

Task 2: Sine and Cosine Plot

    Task: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.


```python
x = np.linspace(0, 2*np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='$\sin(x)$', linestyle='-', marker='o', color='b')
plt.plot(x, y2, label='$\cos(x)$', linestyle='--', marker='x', color='r')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $\sin(x)$ and $\cos(x)$')
plt.legend()

plt.grid(True)
plt.show()
```

Task 3:Subplots

    Task: Create a 2x2 grid of subplots. In each subplot, plot:
        Top-left: $ f(x) = x^3 $
        Top-right: $ f(x) = \sin(x) $
        Bottom-left: $ f(x) = e^x $
        Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)

    Customize each plot with titles, axis labels, and different colors.


```python
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(0, 2*np.pi, 400)
x3 = np.linspace(0, 10, 400)
x4 = np.linspace(0, 10, 400)

y1 = x1**3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x1, y1, color='b')
axs[0, 0].set_title('$f(x) = x^3$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')


axs[0, 1].plot(x2, y2, color='r')
axs[0, 1].set_title('$f(x) = \sin(x)$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')


axs[1, 0].plot(x3, y3, color='g')
axs[1, 0].set_title('$f(x) = e^x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')


axs[1, 1].plot(x4, y4, color='m')
axs[1, 1].set_title('$f(x) = \log(x+1)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')

plt.tight_layout()
plt.show()
```

Task 4: Scatter Plot


    Task: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.

Task 5: Histogram

    Task: Generate a random dataset of 1000 values sampled from a normal distribution (mean=0, std=1). Plot a histogram of the data with 30 bins. Add a title and axis labels. Adjust the transparency of the bars using the alpha parameter.


```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, alpha=0.75)

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Values from a Normal Distribution')

plt.grid(True)
plt.show()
```

Task 6: 3D Plotting

    Task: Create a 3D surface plot for the function $ f(x, y) = \cos(x^2 + y^2) $ over the range of $ x $ and $ y $ values from -5 to 5. Use a suitable colormap and add a colorbar. Set appropriate labels for the axes and title.


```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.cos(x**2 + y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


surf = ax.plot_surface(x, y, z, cmap='viridis')

fig.colorbar(surf)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of $f(x, y) = \cos(x^2 + y^2)$')


plt.show()
```

Task 7: Bar Chart

Task 8: Stacked Bar Chart

    Task: Create a stacked bar chart that shows the contribution of three different categories ('Category A', 'Category B', and 'Category C') over four time periods ('T1', 'T2', 'T3', 'T4'). Use sample data for each category at each time period. Customize the chart with a title, axis labels, and a legend.


```python
categories = ['Category A', 'Category B', 'Category C']
time_periods = ['T1', 'T2', 'T3', 'T4']
data = {
    'Category A': [3, 4, 2, 5],
    'Category B': [2, 3, 4, 1],
    'Category C': [1, 2, 3, 4]
}

category_a = np.array(data['Category A'])
category_b = np.array(data['Category B'])
category_c = np.array(data['Category C'])

fig, ax = plt.subplots()

bar_width = 0.5
index = np.arange(len(time_periods))

bar1 = ax.bar(index, category_a, bar_width, label='Category A')
bar2 = ax.bar(index, category_b, bar_width, bottom=category_a, label='Category B')
bar3 = ax.bar(index, category_c, bar_width, bottom=category_a + category_b, label='Category C')

ax.set_xlabel('Time Periods')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart of Categories over Time Periods')
ax.set_xticks(index)
ax.set_xticklabels(time_periods)
ax.legend()

plt.show()
```
